```python
import pandas as pd
```

```python
data_path = "../data/202302_YouBike_Rental.csv"
df = pd.read_csv(data_path)
```

```python
# Rename columns to English
df.rename(columns={
    "借車時間": "Borrow Time",
    "借車站": "Borrow Station",
    "還車時間": "Return Time",
    "還車站": "Return Station",
    "租借時數": "Rental Hours",
    "借車日期": "Borrow Date"
}, inplace=True)

print(df)
```

```python
from utils import fetch_stations

stations = fetch_stations('../youbike_data.db')
stations.head()
```

```python
station_names = [x.split('_')[1] for x in stations.sna.to_list()]
```

```python
# Create a new column by splitting and extracting the part after '_'
stations['Borrow Station'] = stations['sna'].apply(lambda x: x.split('_')[1])
stations['Return Station'] = stations['sna'].apply(lambda x: x.split('_')[1])


# Merge the DataFrames on the new column
merged_df = pd.merge(stations, df, on='Borrow Station', how='inner')
merged_df_return = pd.merge(stations, df, on='Retrun Station', how='inner')
```

```python
merged_df.head()
```

```python
# Count occurrences of each value in the "Station" column to check how often a bike has been borrowed from a station
value_counts = merged_df["snaen"].value_counts()

print(value_counts)
```

```python
import matplotlib.pyplot as plt

# Plot the value counts as a bar chart
plt.figure(figsize=(24, 18))
value_counts[:100].plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Station Usage Counts', fontsize=16)
plt.xlabel('Station', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=90, fontsize=14)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```
