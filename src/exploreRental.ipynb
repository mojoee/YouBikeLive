```python
import pandas as pd
```

```python
data_path = "../data/202302_YouBike_Rental.csv"
df = pd.read_csv(data_path)
```

```python
# Rename columns to English
df.rename(columns={
    "借車時間": "Borrow Time",
    "借車站": "Borrow Station",
    "還車時間": "Return Time",
    "還車站": "Return Station",
    "租借時數": "Rental Hours",
    "借車日期": "Borrow Date"
}, inplace=True)

print(df)
```

```python
# Create a new column by splitting and extracting the part after '_'
stations['Borrow Station'] = stations['sna'].apply(lambda x: x.split('_')[1])
stations['Return Station'] = stations['sna'].apply(lambda x: x.split('_')[1])


# Merge the DataFrames on the new column
merged_df = pd.merge(stations, df, on='Borrow Station', how='inner')
merged_df_return = pd.merge(stations, df, on='Return Station', how='inner')
```

```python
merged_df.head()
```

```python
# Count occurrences of each value in the "Station" column to check how often a bike has been borrowed from a station
value_counts = merged_df["snaen"].value_counts()

print(value_counts)
```

```python
import matplotlib.pyplot as plt

# Plot the value counts as a bar chart
plt.figure(figsize=(24, 18))
value_counts[:100].plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Station Usage Counts', fontsize=16)
plt.xlabel('Station', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=90, fontsize=14)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Convert Timestamp to datetime if it's not already
df["Timestamp"] = pd.to_datetime(df["Borrow Time"])

# Resample to daily frequency for aggregation
df_daily = df.resample("D", on="Timestamp").sum()

# Group by Borrow Date and count the number of rentals
rental_counts = df.groupby("Borrow Date").size()

# Plot the rental counts over time
plt.figure(figsize=(10, 6))
rental_counts.plot(kind="line", marker="o", linestyle="-")
plt.title("Number of Rentals Over the Month")
plt.xlabel("Date")
plt.ylabel("Number of Rentals")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Convert "Rental Hours" to total seconds
def time_to_seconds(time_str):
    h, m, s = map(int, time_str.split(":"))
    return h * 3600 + m * 60 + s

# Convert "Rental Hours" to total seconds
def time_to_seconds(time_str):
    h, m, s = map(int, time_str.split(":"))
    return h * 3600 + m * 60 + s

df["Trip Duration (seconds)"] = df["Rental Hours"].apply(time_to_seconds)

# Convert seconds to minutes for better readability
df["Trip Duration (minutes)"] = df["Trip Duration (seconds)"] / 60

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(
    df["Trip Duration (minutes)"], 
    bins=30,  # Smaller bin size for more detailed distribution
    range=(0, 60),  # Limit x-axis range to 60 minutes
    color="skyblue", 
    edgecolor="black", 
    alpha=0.7
)
plt.title("Trip Duration Distribution", fontsize=16)
plt.xlabel("Trip Duration (Minutes)", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```
